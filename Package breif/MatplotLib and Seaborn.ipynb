```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel("years")
plt.ylabel("Yeild(Tons per hectare)")
plt.legend(['Apples','Oranges'])
plt.title('Kheti')
```

```python
plt.plot(years,apples,marker='.')
plt.plot(years,oranges,marker='X')
#now all the points are shown as highlighted with markers
```

```python
# fmt='[marker][line][color]'
plt.plot(years,apples,'s-b')
plt.plot(years,oranges,'o--r')
```

```python
#cange the figure
plt.figure(figsize=(10,5))
plt.plot(years,oranges,'or')
```

```python
setosa_df=flowers_df[flowers_df.species=='setosa']
versicolor_df=flowers_df[flowers_df.species=='versicolor']
virginica_df=flowers_df[flowers_df.species=='virginica']
```

```python
setosa_df
```

```python
versicolor_df
```

```python
virginica_df
```

```python
plt.hist(setosa_df.sepal_width,alpha=0.3,bins=np.arange(2,5,0.25))
plt.hist(versicolor_df.sepal_width,alpha=0.3,bins=np.arange(2,5,0.25))
plt.hist(virginica_df.sepal_width,alpha=0.3,bins=np.arange(2,5,0.25))

plt.legend(["Setosa","Versicolor","Virginica"])
```

```python
# we can also stack histogram on one another
plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width],bins=np.arange(2,5,0.25),stacked=True);
```

# Bar chart

```python
years = range(2000, 2006)
apples = [0.35, 0.6, 0.9, 0.8, 0.65, 0.8]
oranges = [0.4, 0.8, 0.9, 0.7, 0.6, 0.8]
```

```python
 plt.bar(years,apples,alpha=0.3);
plt.bar(years,oranges,alpha=0.3);
```

```python
plt.bar(years,oranges)
plt.plot(years,oranges,'o--r')
```

```python
plt.bar(years,apples)
plt.bar(years,oranges,bottom=apples);
```

```python
tips_df=sns.load_dataset('tips')
```

```python
# what is average bill per day

avg_day_df=tips_df.groupby('day')[['total_bill']].mean()
avg_day_df
```

# Heatmap

```python
df=sns.load_dataset('flights')
```

```python
plt.plot(df.passengers)
```

```python
flights_df=sns.load_dataset('flights').pivot('month','year','passengers')
```

```python
flights_df
```

```python
sns.heatmap(flights_df);
plt.title('Passenger no(1000s)');
```

```python
sns.heatmap(flights_df,fmt="d",annot=True,cmap="Blues");
#blues is a type of inbuilt map in seaborn library
```
