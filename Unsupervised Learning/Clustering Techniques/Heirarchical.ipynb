```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
iris_df=pd.read_csv('C:\\Users\\harsh\\Desktop\\iris.csv')
```

```python
iris_df.rename(columns={'150':'pl','4':'pw','setosa':'sl','versicolor':'sw','virginica':'species'},inplace=True)
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage,fcluster
```

```python
heirarchy=linkage(iris_df,method='complete')
```

```python
dendro=dendrogram(heirarchy)
```

```python
labels=fcluster(heirarchy,4,criterion='distance')
```

```python
sns.scatterplot(iris_df.pl,iris_df.pw,hue=labels)
```
